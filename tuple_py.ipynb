```python
# Program to add two matrices using nested loop

X = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]

Y = [[5,8,1],
    [6,7,3],
    [4,5,9]]

result = [[0,0,0],
         [0,0,0],
         [0,0,0]]

# iterate through rows
for i in range(len(X)):
   # iterate through columns
   for j in range(len(X[0])):
       result[i][j] = X[i][j] + Y[i][j]

for r in result:
   print(r)
```

```python
def replace_vowels(string, char):
    vowels = ['a','e','i','o','u']

    for i in string:
        if i in vowels:
          string = string.replace(i,char)
    return string

print(replace_vowels('hello','y'))
```

```python
def replace_vowels(string, char):
    vowels = ['a','e','i','o','u']
    new = string # initialize new with the original string
    for i in string:
        if i in vowels: # check if the character is a vowel
            new = new.replace(i,char) # replace the vowel with the specified character
    return new

print(replace_vowels('hello','y'))
```

```python
def has_consecutive_letters(s):
    for i in s:
         if i == i+1:
          return True
         else:
            return False
print(has_consecutive_letters('abcde'))
```

```python
def has_consecutive_letters(s):
    for i in range(len(s) - 1): # loop through the string, but stop one character short of the end
        if ord(s[i]) == ord(s[i + 1]) - 1: # compare index of current char to index of next char
            return True
    return False # return False if no consecutive letters found

print(has_consecutive_letters('good'))
```

```python
def concatenate_items(items):
    add = 0
    for i in range(len(items)):
        add += str(items[i])
    return add
print(concatenate_items(['hello','word']))
```
