```python
import numpy as np

def my_is_orthogonal(v1,v2,tol):

  A1 = np.linalg.norm(v1)
  A2 = np.linalg.norm(v2)

  # Handle cases where one or both vectors might be zero vectors
  if A1 == 0 or A2 == 0:
    return 1 # Zero vector is considered orthogonal to all vectors

  # Use np.dot for 1D arrays to get scalar dot product
  dot_product = np.dot(v1,v2)

  # To prevent floating point errors causing values slightly outside [-1, 1] for arccos
  cosine_theta = np.clip(dot_product/(A1*A2), -1.0, 1.0)
  theta = np.arccos(cosine_theta)

  # Check for orthogonality (angle close to pi/2)
  if np.abs(theta - np.pi/2) < tol:
    return 1
  else:
    return 0

#Call .flatten() on the numpy array object
v1 = np.array([[1],[0.001]]).flatten()
v2 = np.array([[0.001], [1]]).flatten()

print(my_is_orthogonal(v1, v2, 0.01))
print(v1, v2)
```

```python
import numpy as np

def my_make_ind(A):

  n = np.linalg.matrix_rank(A) # rank of a matrix
  B = []

  for j in range(A.shape[1]):

    #checking column one by one
    col = A[:, j:j+1]

    if len(B)==0:
      B.append(col)

    else:
      #test for independent column
      column_stack = np.hstack([np.hstack(B), col])

      rank_column_stack = np.linalg.matrix_rank(column_stack)

      if rank_column_stack > len(B):
        B.append(col)

    if len(B)==n:
      break
  return np.hstack(B)

A = np.array([[12,24,0,11,-24,18,15],
              [19,38,0,10,-31,25,9],
              [1,2,0,21,-5,3,20],
              [6,12,0,13,-10,8,5],
              [22,44,0,2,-12,17,23]])

B = my_make_ind(A)
print(len(B))
print("Rank of B:", np.linalg.matrix_rank(B))
```

```python
# code to find out matrix is square or not
def sqr_matrix(A):
  num_rows = len(A)

  for row in A:
    num_column = len(row)
    if num_column != num_rows:
      return None
    else: return A

A = [[1,2],
     [3,4]]

B = [[1,2,3], [4,5,6]]

print(sqr_matrix(A))
print(sqr_matrix(B))
```

```python
# creating minor of a matrix using numpy

def create_minor(matrix, row, col):
    """Remove specified row and column."""
    del_row = np.delete(matrix, row, axis=0)
    del_column = np.delete(del_row, col, axis=1 )
    minor = del_column
    return minor

A = np.array([[1,2, 3], [3, 4, 5], [5, 6, 7]])
print(create_minor(A, 0, 0))
```

```python
# minor of a matrix without using numpy
def minor(matrix, row, col):
   for i in matrix.shape[0]:
    for j in matrix.shape[1]:
      if i != row and j!= col:
        minor = np.array([[matrix[i][j]]])
    return minor

minor = np.array([[1,0, 3], [0, 4, 5], [0, 6, -5]])

for row in range(minor.shape[0]):
    for col in range(minor.shape[1]):
         print(f'M[{row}][{col}]:\n',create_minor(minor, row, col))
```

```python
# defining the determinant of a matrix using recurssion relation

def my_rec_det(M):

# checking for square matrix
   if M.shape[0] != M.shape[1]:
      return None

# the determinant of 2d square matrix
   elif M.shape[0] == 2:
      det_m = M[0][0]*M[1][1] - M[1][0]*M[0][1]
      return det_m

# determinant of nd squate matrix
   else:
       det = 0
       for j in range(M.shape[1]):
           minor = create_minor(M,0,j)
           get_det = (-1)**j * M[0][j]*my_rec_det(minor)
           det += get_det
       return det

M = np.array([[1, 2, 11], [-5, 1, 10], [3, 9, 1]])
A = np.array([[1,2], [-5,4]])
B = np.array([[1, 0, -5, -8], [2, 7, 0, 1], [1, -2, 0, -4], [1, 8, 4, 0]])
mat = np.array([[1,2,-9], [3, 0 ,-7]])
print(my_rec_det(M))
print(my_rec_det(A))
print(my_rec_det(B))
print(my_rec_det(mat))
```

```python
from sympy import symbols, Poly

def f(x):
  return x**2+2

def df(x):
  return 2*x

def coeff_vec(p, dp):
  x = symbols('x')

  exp = p(x)
  dexp = dp(x)

  p = Poly(exp,x).all_coeffs()
  D = Poly(dexp,x).all_coeffs()
  return p, D



print(coeff_vec(f, df))
```

```python
# finding largest matrix

import numpy as np

x = np.array([1, 1])
a = np.array([[0, 2], [2, 3]])

tol = 0.01
max_iter = 100

old_eigenval = 0

for i in range(max_iter+1):
  y = np.dot(a,x)

  eigenval = abs(y).max()
  eigenvec = y/eigenval

  if np.abs(eigenval-old_eigenval) < tol:
    break

  x = eigenvec
  old_eigenval = eigenval


  print(f'Iterations: {i+1}')
  print(f'Eigenvalue: {eigenval}')
  print(f'Eigenvector: {eigenvec}')

print(f'eigenvale and eigenvector converged after {i} iteration')
```

```python
# finding smallest eigenvalue

import numpy as np

x = np.array([1, 1])
a = np.array([[0, 2], [2, 3]])

a_inv = np.linalg.inv(a)

tol = 0.01
max_iter = 100

old_eigenval = 0

for i in range(max_iter):
  y = np.dot(a_inv,x)

  eigenval = abs(y).max()
  eigenvec = y/eigenval

  if np.abs(eigenval-old_eigenval) < tol:
    break

  x = eigenvec
  old_eigenval = eigenval


  print(f'Iterations: {i+1}')
  print(f'Eigenvalue: {eigenval}')
  print(f'Eigenvector: {eigenvec}')

print(f'eigenvale and eigenvector converged after {i} iteration')
```
